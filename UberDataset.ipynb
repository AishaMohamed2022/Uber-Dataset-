```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/UberDataset.xlsx')

print(df.head())
```

```python
#missing values in the PURPOSE column with 'Unspecified'
df['PURPOSE'] = df['PURPOSE'].fillna('Unspecified')
print(df)
```

```python
purpose_counts = df['PURPOSE'].value_counts().reset_index()
purpose_counts.columns = ['Purpose', 'Count']
print(purpose_counts)
```

```python
STOP_counts = df['STOP'].value_counts().reset_index()
STOP_counts.columns = ['STOP', 'Count']
START_counts = df['START'].value_counts().reset_index()
START_counts.columns = ['START', 'Count']
START_counts
```

**clean some data to get the right result**

```python
# Right name  R?walpindi to Rawalpindi
df['START'] = df['START'].str.replace(r'R\?walpindi', 'Rawalpindi', regex=True)
df['STOP'] = df['STOP'].str.replace(r'R\?walpindi', 'Rawalpindi', regex=True)
```

```python
#Kar?chi  to, Karachi
df['START'] = df['START'].str.replace(r'Kar\?chi', 'Karachi', regex=True)
df['STOP'] = df['STOP'].str.replace(r'Kar\?chi', 'Karachi', regex=True)
```

```python
print(df['START'].unique())
print(df['STOP'].unique())
```

**Extract Time , Day and Month from datetime column**

```python
df['START_trip'] = pd.to_datetime(df['START_DATE']).dt.time
df['Month'] = pd.to_datetime(df['START_DATE']).dt.month_name()
df['Day'] = pd.to_datetime(df['START_DATE']).dt.day_name()
df['END_trip'] = pd.to_datetime(df['END_DATE']).dt.time

print(df['END_trip'])
```

```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'])
df['END_DATE'] = pd.to_datetime(df['END_DATE'])
```

**Calculate duration for each trip**

```python
df['Duration_trip'] = (df['END_trip'] - df['START_trip'])
```

**Add a New column to classify the time of day**

```python
def time_of_day(hour):
    if hour < 12:
        return "Morning"
    elif hour < 17:
        return "Afternoon"
    elif hour < 21:
        return "Evening"
    else:
        return "Night"

df['TimeOfDay'] = pd.to_datetime(df['START_DATE']).dt.hour.apply(time_of_day)
print(df['TimeOfDay'].head(5))
```
